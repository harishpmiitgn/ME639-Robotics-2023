```python
import numpy as np

def Dh(d, theta, alpha, a):
    """
    Function to calculate the Denavit-Hartenberg matrix for a joint.

    Parameters:
    - d: Link offset
    - theta: Joint angle
    - alpha: Link twist
    - a: Link length

    Returns:
    - dh: Denavit-Hartenberg matrix
    """
    dh = np.zeros((4, 4))
    dh[0][0] = np.cos(theta)
    dh[0][1] = -np.sin(theta) * np.cos(alpha)
    dh[0][2] = np.sin(theta) * np.sin(alpha)
    dh[0][3] = a * np.cos(theta)
    dh[1][0] = np.sin(theta)
    dh[1][1] = np.cos(theta) * np.cos(alpha)
    dh[1][2] = -np.cos(theta) * np.sin(alpha)
    dh[1][3] = a * np.sin(theta)
    dh[2][1] = np.sin(alpha)
    dh[2][2] = np.cos(alpha)
    dh[2][3] = d
    dh[3][3] = 1
    return dh

def inverse_kinematics(x, y, z):
    """
    Function to calculate the inverse kinematics for a SCARA manipulator.

    Parameters:
    - x, y, z: End-effector coordinates

    Returns:
    - theta1, theta2, d3: Joint angles
    """
    theta2 = np.arccos((x**2 + y**2 - l1**2 - l2**2) / (2 * l1 * l2))
    theta1 = np.arctan2(y, x) - np.arctan2(l2 * np.sin(theta2), (l1 + l2 * np.cos(theta2)))
    d3 = -z
    return theta1, theta2, d3

def forward_kinematics(q):
    """
    Function to calculate the forward kinematics for a SCARA manipulator.

    Parameters:
    - q: Joint angles

    Returns:
    - x, y, z: End-effector coordinates
    """
    theta1, theta2, d3 = q
    T01 = Dh(0, theta1, 0, l1)
    T12 = Dh(0, theta2, np.pi, l2)
    T23 = Dh(d3, 0, 0, 0)

    T02 = np.matmul(T01, T12)
    T03 = np.matmul(T02, T23)
    x, y, z = T03[0:3, 3]
    return x, y, z

# Manipulator parameters
l1 = 0.25  # length of link in meters
l2 = 0.25  # length of link in meters
l3 = 0.25  # length of link in meters

# Points A, B, C, and D
A = [0.4, 0.06, 0.1]
B = [0.4, 0.01, 0.1]
C = [0.35, 0.01, 0.1]
D = [0.35, 0.06, 0.1]

# Calculate and print forward kinematics for each point
print("For A:", forward_kinematics(inverse_kinematics(A[0], A[1], A[2])))
print("For B:", forward_kinematics(inverse_kinematics(B[0], B[1], B[2])))
print("For C:", forward_kinematics(inverse_kinematics(C[0], C[1], C[2])))
print("For D:", forward_kinematics(inverse_kinematics(D[0], D[1], D[2])))
```

```python
import numpy as np
import matplotlib.pyplot as plt

def inverse_kinematics(x, y, z):
    theta2 = np.arccos((x**2 + y**2 - l1**2 - l2**2) / (2 * l1 * l2))
    theta1 = np.arctan2(y, x) - np.arctan2(l2 * np.sin(theta2), (l1 + l2 * np.cos(theta2)))
    d3 = -z
    return theta1, theta2, d3

l1 = 0.25  # length of link in meters
l3 = 0.25  # length of link in meters (Note: l3 was defined but not used in the code)
l2 = 0.25  # length of link in meters

A = [0.4, 0.06, 0.1]
B = [0.4, 0.01, 0.1]
C = [0.35, 0.01, 0.1]
D = [0.35, 0.06, 0.1]

points = np.array([A, B, C, D])
timetaken = 100
t_smol = np.linspace(0, timetaken, timetaken)

def genlinear_trajectory(initial, final, t):
    traj = initial + (final - initial) * t / timetaken
    return traj

theta1 = []
theta2 = []
d3 = []
xe = []
ye = []
ze = []

for i in range(4):
    for t in t_smol:
        x, y, z = genlinear_trajectory(points[i, :], points[(i + 1) % 4, :], t)
        xe.append(x)
        ye.append(y)
        ze.append(z)
        q = inverse_kinematics(x, y, z)
        theta1.append(q[0])
        theta2.append(q[1])
        d3.append(q[2])

tnet = np.linspace(0, 4 * timetaken, 4 * timetaken)

fig, axs = plt.subplots(2, 3, figsize=(15, 10), sharex=True)

axs[0, 0].plot(tnet, xe, label='xe')
axs[0, 1].plot(tnet, ye, label='ye')
axs[0, 2].plot(tnet, ze, label='ze')
axs[1, 0].plot(tnet, np.rad2deg(theta1), label='theta1')
axs[1, 1].plot(tnet, np.rad2deg(theta2), label='theta2')
axs[1, 2].plot(tnet, d3, label='d3')

axs[1, 0].set_xlabel('Time (arbitrary units)')
axs[1, 1].set_xlabel('Time (arbitrary units)')
axs[1, 2].set_xlabel('Time (arbitrary units)')

axs[0, 0].set_ylabel('X position')
axs[0, 1].set_ylabel('Y position')
axs[0, 2].set_ylabel('Z position')
axs[1, 0].set_ylabel('Theta1 (degrees)')
axs[1, 1].set_ylabel('Theta2 (degrees)')
axs[1, 2].set_ylabel('D3 position')

fig.suptitle('Modified SCARA Robot Trajectory')

for ax_row in axs:
    for ax in ax_row:
        ax.legend()

plt.show()
```
